```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#orders until 30th of November
train = pd.read_csv("data/training_set_v2.csv", )
train = train.drop('Unnamed: 0',axis=1)
#December set
test_dec = pd.read_csv("data/testset_dec_one_hot.csv")
test_dec = test_dec.drop('Unnamed: 0',axis=1)
#Jan set
test_jan = pd.read_csv("data/testset_jan_one_hot.csv")
test_jan = test_jan.drop('Unnamed: 0',axis=1)

display(test_dec.head())
```

```python
#preprocess train to have dates-> int and segement as onehot
# only train needed because of mistake in train test notebook (wrong dataframe was used)

#transform columns into dates
#trainset
train['date'] = pd.to_datetime(train['date'])
train['MinPurchaseDate'] = pd.to_datetime(train['MinPurchaseDate'])
train['MaxPurchaseDate']= pd.to_datetime(train['MaxPurchaseDate'])




#get min Date to start the count of days as integers
fromDate = min(train['date'])

train['date'] = (train['date'] - fromDate).dt.days.astype(int)
train['MinPurchaseDate'] = (train['MinPurchaseDate'] - fromDate).dt.days.astype(int)
train['MaxPurchaseDate']= (train['MaxPurchaseDate'] - fromDate).dt.days.astype(int)

#one hot encoder for segment
one_hot = pd.get_dummies(train['Segment'])
train = train.drop('Segment',axis = 1)
train = train.join(one_hot)

train.head()
```

```python
#preprocessing for DEC
#testset Dec
test_dec['date'] = pd.to_datetime(test_dec['date'])
test_dec['MinPurchaseDate'] = pd.to_datetime(test_dec['MinPurchaseDate'])
test_dec['MaxPurchaseDate']= pd.to_datetime(test_dec['MaxPurchaseDate'])

#transform dates into count
test_dec['date'] = (test_dec['date'] - fromDate).dt.days.astype(int)
test_dec['MinPurchaseDate'] = (test_dec['MinPurchaseDate'] - fromDate).dt.days.astype(int)
test_dec['MaxPurchaseDate']= (test_dec['MaxPurchaseDate'] - fromDate).dt.days.astype(int)

display(test_dec)
```

```python
#preprocessing for JAN
#testset Jan
test_jan['date'] = pd.to_datetime(test_jan['date'])
test_jan['MinPurchaseDate'] = pd.to_datetime(test_jan['MinPurchaseDate'])
test_jan['MaxPurchaseDate']= pd.to_datetime(test_jan['MaxPurchaseDate'])
#transform dates into count
test_jan['date'] = (test_jan['date'] - fromDate).dt.days.astype(int)
test_jan['MinPurchaseDate'] = (test_jan['MinPurchaseDate'] - fromDate).dt.days.astype(int)
test_jan['MaxPurchaseDate']= (test_jan['MaxPurchaseDate'] - fromDate).dt.days.astype(int)

test_jan
```

```python
def count_points(pred, gold):
    df = pd.merge(pred, gold, on=['userID', 'itemID'], suffixes=('_pred', '_gold'))
    df['points'] = df.apply(_compute_points_for_row, axis=1)
    print(df.shape)
    return df['points'].sum()

def _compute_points_for_row(row):
    y_pred, y_gold = row.prediction_pred, row.prediction_gold
    if y_pred == y_gold:
        # one point if "no order" (0) is predicted correctly; three points if order week is predicted correctly
        return 1 if y_pred == 0 else 3
    # one point if order is predicted correctly (but not the correct week), otherwise zero points
    return 1 if (y_pred > 0 and y_gold > 0) else 0
```
